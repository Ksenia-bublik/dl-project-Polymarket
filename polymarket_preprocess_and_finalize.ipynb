```python
from pathlib import Path
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1200)
pd.set_option("display.max_rows", 200)
```

```python
INPUT_CSV_PATH = r"D:\ДЛ\data\polymarket_culture_2024_to_now_closed_resolved\culture_markets_2024_to_now_closed_resolved_yesno.csv"
input_path = Path(INPUT_CSV_PATH)
print("INPUT_CSV_PATH =", input_path)
print("Exists:", input_path.exists())
```

```python
if not input_path.exists():
    raise FileNotFoundError(f"Файл не найден: {input_path}")

df = pd.read_csv(input_path)
print("Shape:", df.shape)
display(df.head(5))
```

```python
# базовая инфа
print("Rows, cols:", df.shape)
print("\nColumns:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:>3}. {col}")

print("\nDtypes:")
display(df.dtypes.to_frame("dtype"))
```

```python
# пропуски
missing_summary = pd.DataFrame({
    "missing_count": df.isna().sum(),
    "missing_pct": (df.isna().mean() * 100).round(2),
}).sort_values(["missing_count", "missing_pct"], ascending=False)

display(missing_summary)
```

```python
# поиск дубликатов
exact_duplicates = df.duplicated().sum()
print("Полных дубликатов строк:", exact_duplicates)

dup_checks = {}

for col in ["market_id", "condition_id", "question", "slug"]:
    if col in df.columns:
        dup_checks[col] = int(df[col].duplicated().sum())

print("\nДубликаты по ключевым колонкам:")
for k, v in dup_checks.items():
    print(f"{k}: {v}")

for col in ["market_id", "condition_id"]:
    if col in df.columns:
        dup_rows = df[df[col].duplicated(keep=False)].sort_values(col)
        print(f"\nСтроки с дубликатами по {col}: {len(dup_rows)}")
        if len(dup_rows) > 0:
            display(dup_rows.head(20))
```

```python
# выполнение бинарности ответов
if "is_binary_yes_no" not in df.columns:
    print("Колонка is_binary_yes_no отсутствует.")
    bad_is_binary = pd.DataFrame()
else:
    mask_bad_is_binary = df["is_binary_yes_no"].fillna(False) != True
    bad_is_binary = df.loc[mask_bad_is_binary].copy()
    print("Количество строк, где is_binary_yes_no != True:", len(bad_is_binary))
    display(bad_is_binary.head(50))
```

```python
if "token_count" not in df.columns:
    print("Колонка token_count отсутствует.")
    bad_token_count = pd.DataFrame()
else:
    token_count_numeric = pd.to_numeric(df["token_count"], errors="coerce")
    mask_bad_token_count = token_count_numeric != 2
    bad_token_count = df.loc[mask_bad_token_count].copy()
    print("Количество строк, где token_count != 2:", len(bad_token_count))
    display(bad_token_count.head(50))

    print("\nРаспределение token_count:")
    display(token_count_numeric.value_counts(dropna=False).sort_index())
```

```python
if "umaResolutionStatus" not in df.columns:
    print("Колонка umaResolutionStatus отсутствует.")
    status_counts = pd.Series(dtype="int64")
else:
    status_series = (
        df["umaResolutionStatus"]
        .astype("string")
        .str.strip()
        .str.lower()
    )
    status_counts = status_series.fillna("<NA>").value_counts(dropna=False)

    print("Распределение umaResolutionStatus:")
    display(status_counts.to_frame("count"))

    resolved_count = int((status_series == "resolved").sum())
    proposed_count = int((status_series == "proposed").sum())
    other_values = sorted(set(status_series.dropna().unique()) - {"resolved", "proposed"})

    print("resolved:", resolved_count)
    print("proposed:", proposed_count)
    print("other values:", other_values if other_values else "нет")
```

```python
#удаление лишнего
COLUMNS_TO_DROP = [
    "market_id",
    "condition_id",
    "slug",
    "category",
    "updatedAt",
    "closedTime",
    "active",
    "closed",
    "archived",
    "acceptingOrders",
    "marketType",
    "formatType",
    "liquidity",
    "volume",
    "liquidityNum",
    "volumeNum",
    "bestBid",
    "bestAsk",
    "lastTradePrice",
    "spread",
    "outcomes_raw",
    "outcomePrices_raw",
    "tag_labels",
    "tag_slugs",
    "tags_json",
    "categories_json",
    "event_id",
    "event_slug",
    "event_title",
    "event_category",
    "event_subcategory",
    "event_creationDate",
    "event_endDate",
    "event_createdAt",
    "event_updatedAt",
    "createdAt_dt",
    "active_simplified",
    "closed_simplified",
    "archived_simplified",
    "accepting_orders_simplified",
    "tokens_json",
    "resolution_timestamp_best_effort",
    "is_binary_yes_no",
    "token_count",
    "createdAt"
]

existing_to_drop = [c for c in COLUMNS_TO_DROP if c in df.columns]
missing_to_drop = [c for c in COLUMNS_TO_DROP if c not in df.columns]

print("Будут удалены колонки:", len(existing_to_drop))
display(pd.DataFrame({"drop_column": existing_to_drop}))

print("\nНе найдены в таблице:", len(missing_to_drop))
display(pd.DataFrame({"missing_column": missing_to_drop}))
```

```python
final_df = df.drop(columns=existing_to_drop, errors="ignore").copy()

print("Shape исходной таблицы:", df.shape)
print("Shape final_df после удаления колонок:", final_df.shape)

print("\nОставшиеся колонки:")
for i, col in enumerate(final_df.columns, 1):
    print(f"{i:>3}. {col}")

display(final_df.head(10))
```

```python
final_missing_summary = pd.DataFrame({
    "missing_count": final_df.isna().sum(),
    "missing_pct": (final_df.isna().mean() * 100).round(2),
}).sort_values(["missing_count", "missing_pct"], ascending=False)

display(final_missing_summary)
```

```python
bad_rows_mask = (
    final_df["umaResolutionStatus"].astype("string").str.strip().str.lower().eq("proposed")
    | final_df["endDate"].isna()
    | final_df["startDate"].isna()
)

print("Строк к удалению:", int(bad_rows_mask.sum()))

final_df = final_df.loc[~bad_rows_mask].copy()

print("Новый размер final_df:", final_df.shape)
```

```python
question_dup_mask = final_df["question"].astype("string").str.strip().duplicated(keep=False)

dup_questions_df = final_df.loc[question_dup_mask].copy()

print("Количество строк, у которых question совпадает с другой строкой:", len(dup_questions_df))
print("Количество уникальных текстов question, которые повторяются:", dup_questions_df["question"].nunique())

dup_counts = (
    dup_questions_df["question"]
    .value_counts()
    .rename_axis("question")
    .reset_index(name="count")
)

print("\nТоп-3 повторяющихся question:")
display(dup_counts.head(3))

top3_questions = dup_counts.head(3)["question"].tolist()

sample_cols = [c for c in [
    "question",
    "createdAt",
    "endDate",
    "umaResolutionStatus",
    "winner_outcome",
    "final_outcome"
] if c in final_df.columns]

print("\nТри примера повторяющихся событий:")
display(
    final_df.loc[final_df["question"].isin(top3_questions), sample_cols]
    .sort_values(["question"] + ([c for c in ["createdAt"] if c in sample_cols]))
)
```

```python
final_df = final_df.drop(columns=["umaResolutionStatus", "winner_outcome"], errors="ignore").copy()
final_df.head(5)
```

## 13) Экспорт в CSV и Excel

Сохраняем рядом с исходным файлом.

Имя будет таким:
- `исходное_имя_FINAL.csv`
- `исходное_имя_FINAL.xlsx`

```python
base_stem = input_path.stem
output_dir = input_path.parent

csv_out = output_dir / f"{base_stem}_FINAL.csv"
xlsx_out = output_dir / f"{base_stem}_FINAL.xlsx"

final_df.to_csv(csv_out, index=False)

try:
    final_df.to_excel(xlsx_out, index=False)
    excel_status = "Excel сохранён успешно."
except Exception as e:
    excel_status = f"Ошибка при сохранении Excel: {e}"

print("CSV:", csv_out)
print("XLSX:", xlsx_out)
print(excel_status)
```
